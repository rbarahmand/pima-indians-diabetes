# Setup

```python
# common imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# saving figs automaticaly
PROJECT_ROOT_DIR = "."
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images")
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="jpg", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

# Getting Data Ready

## Import Data

```python
diabetes = pd.read_csv("dataset/clean_out.csv")
diabetes.head(20)
```

```python
diabetes = diabetes.drop("Unnamed: 0", axis=1)
```

```python
diabetes.head()
```

```python
diabetes.info()
```

## Changing Some Data Types

```python
diabetes["Pregnancies"] = diabetes["Pregnancies"].astype(np.uint8)
```

```python
diabetes["Outcome"] = diabetes["Outcome"].astype(np.uint8)
```

```python
diabetes.info()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(diabetes)
diabetes_scaled_array = scaler.transform(diabetes)
```

```python
diabetes_scaled = pd.DataFrame(diabetes_scaled_array, columns=list(diabetes.columns))
diabetes_scaled.head()
```

## Separating Features from Target

```python
X, y = diabetes_scaled.iloc[:,:8], np.ravel(diabetes_scaled.iloc[:,8:])
```

# Getting the Evaluation Pipeline Ready

**Our Problem:** Classification Problem Specially (Unbalanced)
**Evaluation Metrics:**
- Precision
- Recall
- F1
- F2
- ROC AUC

**Our Choice:** ROC AUC

## Creating Repeated Stratified KFold Cross Validation

```python
from sklearn.model_selection import RepeatedStratifiedKFold

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
```

# Stochastic Gradient Descent (SGD) with Elastic Net

```python
from sklearn.linear_model import SGDClassifier
EN_sgd_clf = SGDClassifier(random_state=42, penalty="elasticnet", l1_ratio=0.1)
```

## Model Evaluation

```python
from yellowbrick.model_selection import ValidationCurve
```

```python
vc_viz = ValidationCurve(EN_sgd_clf, param_name="alpha", 
                         param_range=np.arange(0, 0.0001, 0.000005), cv=cv,
                         n_jobs=-1, scoring='roc_auc')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 6));
vc_viz.fit(X, y)
vc_viz.poof()
```

# Stochastic Gradient Descent (SGD) with Elastic Net - SMOTE

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_validate
from sklearn.model_selection import GridSearchCV

model = SGDClassifier(random_state=42, penalty="elasticnet", l1_ratio=0.1)
over = SMOTE()
steps = [("over", over), ("model", model)]
pipe = Pipeline(steps)
param_grid = {
    "over__k_neighbors": list(range(1, 10)),
    "model__alpha": list(np.arange(0.000005, 0.0001, 0.000005)),
}
grid = GridSearchCV(pipe, param_grid, cv=cv, scoring='roc_auc', return_train_score=True, n_jobs=-1, error_score='raise')
grid_result = grid.fit(X, y)
```

```python
mean_test_score = grid_result.cv_results_["mean_test_score"]
mean_train_score = grid_result.cv_results_["mean_train_score"]
param_model__alpha =  grid_result.cv_results_["param_model__alpha"]
param_over__k_neighbors = grid_result.cv_results_["param_over__k_neighbors"]
test_scores = pd.DataFrame({
    "mean_test_scores": mean_test_score,
    "mean_train_scores": mean_train_score,
    "param_model__alpha": param_model__alpha,
    "param_over__k_neighbors": param_over__k_neighbors,
})
```

```python
test_scores
```

```python
df_10_max = test_scores.nlargest(10, ["mean_test_scores"])
df_10_max
```

## Max Achieved Score

```python
df_10_max.iloc[0]
```
